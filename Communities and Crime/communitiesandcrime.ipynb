## **Removendo colunas**

No arquivo foi notado muitas colunas com valores faltantes. Diante disso, nos próximos códigos, foi separada colunas relevantes para fazer as análises sobre a variável 'ViolentCrimesPerPop'. O foco nesta análise foi para pegar dados referente a raça, idade, população e tamanho da área.

```python
#lista de colunas a serem mantidas
colunasAManter = ['population', 'householdsize',
                   'racepctblack', 'racePctWhite', 'racePctAsian', 'racePctHisp',
                   'agePct12t21', 'agePct12t29', 'agePct16t24',
                   'ViolentCrimesPerPop']

#selecionando apenas as colunas desejadas do dataframe
df = df[colunasAManter]
```

```python
#confirmando as colunas conforme alteração no código da janela anterior
df.head()
```

## **Selecionando as características e divisão dos dados da variável 'ViolentCrimesPerPop'**


Este passo visa selecionar as características (ou variáveis) que serão usadas para treinar o modelo de regressão.

No caso deste trabalho:

Variáveis Independentes (X): Estas são todas as colunas de nosso conjunto de dados, exceto a coluna ViolentCrimesPerPop, como vimos outras colunas no código rodado anteriormente. Será usada essa característica como 'previsão' do número de crimes violentos por população Variável Dependente (y): Esta é a coluna ViolentCrimesPerPop. É a variável que queremos prever com base em outras características que escolhemos para fazer a regressão linear.

```python
# Seleção de características e divisão dos dados
X = df.drop('ViolentCrimesPerPop', axis=1)  # Usando todas as variáveis escolhidas anteriormente, exceto 'ViolentCrimesPerPop' como variáveis independentes
y = df['ViolentCrimesPerPop']  # 'ViolentCrimesPerPop' é a variável dependente
```

## **Divisão dos Dados em treinamento e teste**

A prática da divisão em dividir o conjunto de dados para treino e teste ajudará a avaliar como o modelo irá se comportar com dados que não viu antes.

Utilizamos a função train_test_split da biblioteca sklearn para realizar essa divisão. Conforme especificado na letra b deste exercício, foi dividida aleatoriamente a base de dados de treino (train) com 70% das amostras e de teste (test), com 30%.

```python
#Treinamento de modelo com 70% das amostras e testes de performance com 30%.
#Visualizar como o modelo irá se comportar com dados não vistos antes.
#Random 42 garante que teremos a mesma divisão, útil para visualização dos resultados.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## **Treinamento do Modelo**


```python
#Utilização da classe LinearRegression da biblioteca sklearn
#Inicialização do modelo
model = LinearRegression()
```

```python
#Utilização do método fit para treinar o modelo usando o conjunto de treinamento.
#Ele ajusta o modelo aos dados, "aprendendo" a relação entre as características (X_train) e a variável alvo (y_train).
model.fit(X_train, y_train)
```

```python
# Criando um DataFrame para visualizar melhor os coeficientes
coeff_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
})
print(coeff_df)

# Mostrando também a interceptação
print("\nInterceptação:", model.intercept_)
```

## **Realização e visualização das previsões**
Após o treinamento, é hora de testar o modelo em um conjunto de dados não vistos anteriormente, o conjunto de teste. Isso nos dá uma noção de como o modelo se comportará em dados novos e desconhecidos.

Realizando previsões: Usamos o método predict do modelo para prever os valores da variável alvo no conjunto de teste.

```python
#Teste do modelo de conjunto de dados para verificar como o modelo se comportará em novos dados e desconhecidos.
#Y_pred tem a função de prever valores futuros, a partir de suas características

# Gerando previsões no conjunto de teste
y_pred = model.predict(X_test)

# Gerando previsões no conjunto de teste
y_pred_test = model.predict(X_test)
```

```python
# Importando os módeulos necessários
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Calculando RMSE e MAE para o conjunto de treino
# Usando y_test (os valores reais do conjunto de teste) para comparação
train_mse = mean_squared_error(y_test, y_pred)
train_mae = mean_absolute_error(y_test, y_pred)
train_rmse = np.sqrt(train_mse)
```

```python
# Calculando RMSE e MAE para o conjunto de teste
test_mse = mean_squared_error(y_test, y_pred_test)
test_mae = mean_absolute_error(y_test, y_pred_test)
test_rmse = np.sqrt(test_mse)

# Imprimindo os resultados
print('Treino - RMSE: %.4f' % train_rmse)
print('Treino - MAE: %.4f' % train_mae)
print('Teste - RMSE: %.4f' % test_rmse)
print('Teste - MAE: %.4f' % test_mae)
```

```python
# Tamanho e resolução do gráfico
plt.figure(figsize=(10,6), dpi=100)

# Scatter plot
plt.scatter(y_test, y_pred, alpha=0.5, s=40, edgecolors="w", linewidth=0.5)

# Linha de previsão perfeita
min_val = min(min(y_test), min(y_pred))
max_val = max(max(y_test), max(y_pred))
plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--')

# Labels e título
plt.xlabel('ViolentCrimesPerPop num real', fontsize=14)
plt.ylabel('ViolentCrimesPerPop num previsto', fontsize=14)
plt.title('ViolentCrimesPerPop num real vs. num previsto', fontsize=16)

# Mostrar gráfico
plt.tight_layout()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Importando bibliotecas necessárias para a #letra c da questão 1
from sklearn.decomposition import PCA
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
import matplotlib.pyplot as plt

#aplicação do PCA para reduzir para 5 atributos
pca = PCA(n_components=5)  # PCA com 5 atributos principais
X_train_pca = pca.fit_transform(X_train) # PCA de treino
X_test_pca = pca.transform(X_test) #PCA de teste
```

```python
# Visualização dos componentes principais
plt.figure(figsize=(10, 7))
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap='viridis', edgecolor='k', s=40)
plt.title('PCA: principais componentes do conjunto de treino')
plt.xlabel('ViolentCrimesPerPop num real')
plt.ylabel('ViolentCrimesPerPop num previsto')
plt.colorbar(label='valor de y')
plt.show()
```

```python
# Transformação polinomial, como forma p melhorar o modelo
poliPCA = PolynomialFeatures(degree=2)  # grau do polinômio em 2
X_train_poli = poliPCA.fit_transform(X_train_pca) #PCA poli treino
X_test_poli = poliPCA.transform(X_test_pca) #PCA poli teste

#utilizando o fit para treinar o modelo
model_poli = LinearRegression()
model_poli.fit(X_train_poli, y_train)
```

```python
#previsões do modelo
y_pred_train_poli = model_poli.predict(X_train_poli)
y_pred_test_poli = model_poli.predict(X_test_poli)
```

```python
#obtendo os valores de RMSE e MAE sobre o conjunto de treino
train_mse_poli = mean_squared_error(y_train, y_pred_train_poli)
train_mae_poli = mean_absolute_error(y_train, y_pred_train_poli)
train_rmse_poli = np.sqrt(train_mse_poli)

#obtendo os valores de RMSE e MAE sobre o conjunto de test
test_mse_poli = mean_squared_error(y_test, y_pred_test_poli)
test_mae_poli = mean_absolute_error(y_test, y_pred_test_poli)
test_rmse_poli = np.sqrt(test_mse_poli)
#print dos resultados
print('Treino (PCA + polinomial) - RMSE: %.4f' % train_rmse_poli)
print('Treino (PCA + polinomial) - MAE : %.4f' % train_mae_poli)
print('Teste(PCA + polinomial) - RMSE: %.4f' % test_rmse_poli)
print('Teste (PCA + polinomial) - MAE: %.4f' % test_mae_poli)
```
